# Zadanie 1

```python
# Sprawdzanie czy plik "ceny_biletow.txt" istnieje
try:
    # Otwarcie pliku "ceny_biletow.txt" w trybie do odczytu ("r")
    with open("ceny_biletow.txt", "r") as plik:
        bilety = {}  # Tworzenie pustego słownika na ceny biletów
        # Iteracja po każdej linii w pliku
        for linia in plik:
            # Rozdzielanie linii na nazwę biletu i cenę, usuwanie zbędnych znaków i rozdzielanie po przecinku
            nazwa, cena = linia.strip().split(',')
            # Dodawanie do słownika 'bilety' nazwy biletu jako klucza i ceny jako wartości
            bilety[nazwa.strip()] = float(cena.strip())
except FileNotFoundError:
    print("Nie można znaleźć pliku z cenami biletów.")

# Wyświetlanie informacji o dostępnych bileta
print("Dostępne bilety:")
i = 1
for nazwa, cena in bilety.items():
    print(f"{i}. {nazwa}: {cena} PLN")
    i += 1

# Inicjalizacja sumy koszyka oraz pustego słownika koszyka
suma = 0
koszyk = {}

# Pętla nieskończona pozwalająca użytkownikowi dodawać bilety do koszyka
while True:
    wybor = input("Wybierz numer biletu (wpisz jego numer) lub wpisz 'q' aby przejść do płatności: ")

    if wybor == 'q':
        break
        
    # Pobieranie nazwy biletu na podstawie wyboru użytkownika
    nazwa_biletu = list(bilety.keys())[int(wybor) - 1]

    # Pobieranie ilości biletów od użytkownika oraz dodawanie ich do koszyka
    cena_biletu = bilety[nazwa_biletu]

    # Pytanie użytkownika, czy bilet ma być ulgowy
    ulgowy = input(f"Czy bilet '{nazwa_biletu}' ma być ulgowy? (T/N): ")
    if ulgowy.upper() == 'T':
        cena_biletu *= 0.5  # Rabat 50% dla biletu ulgowego

    ilosc = int(input(f"Ile sztuk biletu '{nazwa_biletu}' chcesz dodać do koszyka? "))
    koszyk[nazwa_biletu] = koszyk.get(nazwa_biletu, 0) + ilosc
    suma += cena_biletu * ilosc

# Wyświetlanie łącznej ceny biletów w koszyku
print(f"Łączna cena biletów w koszyku: {suma} PLN")
print(koszyk)

# Pobieranie kwoty płatności od użytkownika i wyliczanie reszty
platnosc = float(input("Podaj kwotę, jaką płacisz: "))
reszta = platnosc - suma

# Sprawdzanie, czy kwota płatności jest wystarczająca
if reszta >= 0:
    print(f"Twoja reszta: {reszta} PLN")
else:
    print("Podana kwota jest za mała.")
```

# Zadanie 2

```python
# Sprawdzanie czy plik "ceny_kawy.txt" istnieje
try:
    # Wczytanie cennika kawy z pliku
    with open("ceny_kawy.txt", "r") as file:
        cennik_kawy = {}
        for line in file:
            nazwa, cena = line.strip().split(',')
            cennik_kawy[nazwa.strip()] = float(cena.strip())
except FileNotFoundError:
    print("Nie można znaleźć pliku z cennikiem kawy.")
    exit()

# Wyświetlenie cennika kawy
print("Dostępne rodzaje kawy:")
i = 1
for nazwa, cena in list(cennik_kawy.items()):
    print(f"{i}. {nazwa}: {cena} PLN")
    i += 1
print("Mleko: 0.01 PLN/ml")
print("Cukier: 0.05 PLN/kostka")

# Inicjalizacja sumy cen wszystkich produktów w koszyku
calkowita_cena = 0

# Inicjalizacja pustego koszyka
koszyk = {}

# Wybór kawy przez użytkownika
while True:
    wybor_kawy = input("Wybierz numer kawy (wpisz jego numer) lub 'q' aby przejść do płatności: ")
    
    if wybor_kawy == 'q':
        break

    # Pobieranie ilości mleka i cukru
    ilosc_mleka = float(input("Podaj ilość mleka (w ml): "))
    ilosc_cukru = float(input("Podaj ilość cukru (w kostkach): "))

    # Dodanie wybranej kawy do koszyka
    nazwa_kawy = list(cennik_kawy.keys())[int(wybor_kawy) - 1]
    
    # Pytanie użytkownika o posiadanie własnego kubka
    posiada_kubek = input(f"Czy posiadasz własny kubek dla kawy {nazwa_kawy}? (t/n): ").lower()

    cena_kawy = cennik_kawy[nazwa_kawy]

    # Obniżenie ceny o 2 zł, jeśli użytkownik posiada kubek
    if posiada_kubek == 't':
        cena_kawy -= 2.0

    calkowita_cena += cena_kawy

    # Dodanie ceny mleka i cukru do całkowitej ceny
    calkowita_cena += ilosc_mleka * 0.01 + ilosc_cukru * 0.05

    koszyk[nazwa_kawy] = koszyk.get(nazwa_kawy, {"ilość": 0, "mleko": 0, "cukier": 0})
    koszyk[nazwa_kawy]["ilość"] += 1
    koszyk[nazwa_kawy]["mleko"] += ilosc_mleka
    koszyk[nazwa_kawy]["cukier"] += ilosc_cukru

# Wyświetlenie całkowitej ceny przed zapłatą
print(f"Cena za cały koszyk: {round(calkowita_cena,2)} PLN")

# Wyświetlenie zawartości koszyka
print("Zawartość koszyka:")
for produkt, szczegoly in koszyk.items():
    print(f"{produkt}: {szczegoly['ilość']} szt., Mleko: {szczegoly['mleko']} ml, Cukier: {szczegoly['cukier']} szt.")

# Pobranie opłaty od użytkownika
oplata = float(input("Podaj kwotę, jaką płacisz: "))

# Obliczenie reszty
reszta = round(oplata - calkowita_cena, 2)
if reszta >= 0:
    print(f"Oto Twoje zamówienie. Reszta wynosi: {reszta} PLN.")
else:
    print("Podana kwota jest za mała.")
```
